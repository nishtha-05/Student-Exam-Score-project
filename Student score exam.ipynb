```python
df = pd.read_csv('Student exam score.csv')
print(df)
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

# Gender Distributation

```python
plt.figure(figsize=(3,3.9))
ax= sns.countplot(x='Gender', data=df ,color='magenta')
plt.title('Gender Distributation')
for bars in ax.containers:
    ax.bar_label(bars)
```

# Impact of parent Education

```python
edu_impact = df.groupby(['ParentEduc']).agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
sns.heatmap(edu_impact , annot=True)
plt.title("Relationship between Parent's Education and Student's score")
plt.show()
```

```python
# from above chart we have concluded that the education of parent have the good impact of children.
```

# impact of parent Marital Status

```python
plt.figure(figsize=(4,3))
gp = df.groupby(['ParentMaritalStatus']).agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
sns.heatmap(gp,annot =True )
plt.title("Relationship between Parent's Maritalstatus and Student's score")
plt.show()
```

```python
#from above chart we have concluded that there is no/negligible impact on  student scores due to of their parent's marital status.
```
